```python
df['discounted_price'] = (
    df['discounted_price']
    .str.replace("₹", '', regex=False)
    .str.replace(",", '', regex=False)
    .astype(float)
)

df['actual_price'] = (
    df['actual_price']
    .str.replace("₹", '', regex=False)
    .str.replace(",", '', regex=False)
    .astype(float)
)
df.actual_price
```

```python
df['discount_percentage'] = (
    df['discount_percentage']
    .str.replace('%', '', regex=False)
    .astype(float)
)
df.discount_percentage
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = (
    df['rating_count']
    .astype(str)
    .str.replace(",", "", regex=False)
    .astype(float)  # or .astype(int) if no decimals
)
df['rating_count'].isnull().sum()
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
```

```python
# creating new dataframe with only useful column for the tableau 
df1 = df[['product_id', 'product_name', 'category', 'discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']].copy()
```

```python
#  Split the category column by "|"
category_split = df1['category'].str.split('|', expand=True)
category_split
```

```python
#  Rename the columns
category_split.columns = [
    'category_level_1',
    'category_level_2',
    'category_level_3',
    'category_level_4',
    'category_level_5',
    'category_level_6',
    'category_level_7' 
]
```

```python
#  Clean each level
for col in category_split.columns:
    category_split[col] = (
        category_split[col]
        .astype(str)                      # Ensure string type
        .str.replace('&', ' & ')          # Add spacing around "&"
        .str.replace('  ', ' ')           # Remove double spaces
        .str.strip()                      # Remove leading/trailing spaces
        .replace('nan', None)             # Clean up empty strings
    )
```

```python
#Join the new columns to your main df1
df1 = pd.concat([df1, category_split], axis=1)
```

```python
#drop the original category column
df1.drop(columns=['category'], inplace=True)
```

```python
 #Checking category_1 unique values

df1['category_level_1'].value_counts()
```

```python
df1['category_level_1'] = df1['category_level_1'].str.replace('OfficeProducts','Office Products')
df1['category_level_1'] = df1['category_level_1'].str.replace('MusicalInstruments ','Musical Instruments')
df1['category_level_1'].value_counts()
```

```python
 #Checking category_2 unique values

df1['category_level_2'].value_counts()
```

```python
df1['category_level_2'] = df1['category_level_2'].str.replace('Kitchen&HomeAppliances', 'Kitchen & Home Appliances')
df1['category_level_2'] = df1['category_level_2'].str.replace('HomeTheater,TV & Video', 'Home Theater, TV & Video')
df1['category_level_2'] = df1['category_level_2'].str.replace('Mobiles & Accessories', 'Mobiles & Accessories')  # already clean
df1['category_level_2'] = df1['category_level_2'].str.replace('Heating,Cooling & AirQuality', 'Heating, Cooling & Air Quality')
df1['category_level_2'] = df1['category_level_2'].str.replace('WearableTechnology', 'Wearable Technology')
df1['category_level_2'] = df1['category_level_2'].str.replace('Headphones,Earbuds & Accessories', 'Headphones, Earbuds & Accessories')
df1['category_level_2'] = df1['category_level_2'].str.replace('NetworkingDevices', 'Networking Devices')
df1['category_level_2'] = df1['category_level_2'].str.replace('OfficePaperProducts', 'Office Paper Products')
df1['category_level_2'] = df1['category_level_2'].str.replace('ExternalDevices & DataStorage', 'External Devices & Data Storage')
df1['category_level_2'] = df1['category_level_2'].str.replace('HomeStorage & Organization', 'Home Storage & Organization')
df1['category_level_2'] = df1['category_level_2'].str.replace('HomeAudio', 'Home Audio')
df1['category_level_2'] = df1['category_level_2'].str.replace('GeneralPurposeBatteries & BatteryChargers', 'General Purpose Batteries & Battery Chargers')
df1['category_level_2'] = df1['category_level_2'].str.replace('Printers,Inks & Accessories', 'Printers, Inks & Accessories')
df1['category_level_2'] = df1['category_level_2'].str.replace('CraftMaterials', 'Craft Materials')
df1['category_level_2'] = df1['category_level_2'].str.replace('OfficeElectronics', 'Office Electronics')
df1['category_level_2'] = df1['category_level_2'].str.replace('PowerAccessories', 'Power Accessories')
df1['category_level_2'] = df1['category_level_2'].str.replace('HomeMedicalSupplies & Equipment', 'Home Medical Supplies & Equipment')
df1['category_level_2'] = df1['category_level_2'].str.replace('CarAccessories', 'Car Accessories')
df1['category_level_2'].value_counts()
```

```python
#Regex Auto-Cleanup for ALL Levels
import re

def insert_spaces(text):
    if pd.isna(text): return text
    return re.sub(r'(?<=[a-z])(?=[A-Z])', ' ', text).replace('  ', ' ').strip()

for col in ['category_level_1', 'category_level_2', 'category_level_3', 'category_level_4', 'category_level_5']:
    df1[col] = df1[col].apply(insert_spaces)
```

```python
#Creating Categories for Rankings

rating_score = []

for score in df1['rating']:
    if score < 2.0 : rating_score.append('Poor')
    elif score < 3.0 : rating_score.append('Below Average')
    elif score < 4.0 : rating_score.append('Average')
    elif score < 5.0 : rating_score.append('Above Average')
    elif score == 5.0 : rating_score.append('Excellent')
```

```python
#Creating A new Column and Changing the Data Type

df1['rating_score'] = rating_score

df1['rating_score'] = df1['rating_score'].astype('category')
```

```python
#Reordered Categories

df1['rating_score'] = df1['rating_score'].cat.reorder_categories(['Below Average', 'Average', 'Above Average', 'Excellent'], ordered=True)
```
